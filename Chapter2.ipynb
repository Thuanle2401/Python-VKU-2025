```python
def encode_caesar(s, d):
    s = list(s)
    alowed = 'abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ'
    for i in range(len(s)):
        if s[i].isalpha():
            if chr(ord(s[i]) + d) in alowed:
                s[i] = chr(ord(s[i]) + d)
            else:
                increase = ord(s[i]) - ord('Z') - 1 + d
                s[i] = chr(ord('A') + increase)
    s = ''.join(s)
    return s

def decode_caesar(s, d):
    s = list(s)
    alowed = 'abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ'
    for i in range(len(s)):
        if s[i].isalpha():
            if chr(ord(s[i]) - d) in alowed:
                s[i] = chr(ord(s[i]) - d)
            else:
                decrease = d - (ord(s[i]) - ord('A') + 1)
                s[i] = chr(ord('Z') - decrease)
    s = ''.join(s)
    return s

s = input("Nhap chuoi cua muon encode caesar: ")
d = int(input("So luong dich chuyen: "))
print("Encode result: " + encode_caesar(s, d))
print("Decode result: " + decode_caesar(encode_caesar(s, d), d))
```

```python
def encode_caesar(s, d):
    result = []
    for ch in s:
        if 'A' <= ch <= 'Z':  
            result.append(chr((ord(ch) - ord('A') + d) % 26 + ord('A')))
        elif 'a' <= ch <= 'z': 
            result.append(chr((ord(ch) - ord('a') + d) % 26 + ord('a')))
        else:
            result.append(ch)  
    return ''.join(result)

def decode_caesar(s, d):
    return encode_caesar(s, -d)  

s = input("Nhap chuoi ban muon encode Caesar: ")
d = int(input("So luong dich chuyen: "))
encoded = encode_caesar(s, d)
decoded = decode_caesar(encoded, d)

print("Encode result:", encoded)
print("Decode result:", decoded)
```

```python
def passFormat(password):
    if len(password) < 8:
        return False
    Upper = 0
    Lower = 0
    Digit = 0
    for c in password:
        if c.isupper():
            Upper += 1
        elif c.isdigit():
            Digit += 1
        elif c.islower():
            Lower += 1
    return (Upper > 0 and Lower > 0 and Digit > 0)

s = input("Type your password to check secure password: ")
if passFormat(s):
    print("Yes, secure !")
else:
    print("No, secure !")
```

```python
import zipfile
import os

def compress_file(input_file, output_zip):
    with zipfile.ZipFile(output_zip, 'w', zipfile.ZIP_DEFLATED) as zipf:
        zipf.write(input_file, os.path.basename(input_file))
    print(f" ÄÃ£ nÃ©n {input_file} thÃ nh {output_zip}")

def decompress_file(zip_file, extract_to):
    with zipfile.ZipFile(zip_file, 'r') as zipf:
        zipf.extractall(extract_to)
    print(f" ÄÃ£ giáº£i nÃ©n {zip_file} vÃ o thÆ° má»¥c {extract_to}")


while True:
    print("\n=== MENU ===")
    print("1. NÃ©n tá»‡p")
    print("2. Giáº£i nÃ©n tá»‡p")
    print("0. ThoÃ¡t")
    choice = input("Chá»n chá»©c nÄƒng: ")

    if choice == "1":
        file_path = input("Nháº­p Ä‘Æ°á»ng dáº«n tá»‡p cáº§n nÃ©n: ")
        output_zip = input("Nháº­p tÃªn file zip Ä‘áº§u ra (vd: file.zip): ")
        compress_file(file_path, output_zip)

    elif choice == "2":
        zip_path = input("Nháº­p Ä‘Æ°á»ng dáº«n file zip cáº§n giáº£i nÃ©n: ")
        extract_dir = input("Nháº­p thÆ° má»¥c lÆ°u file sau khi giáº£i nÃ©n: ")
        decompress_file(zip_path, extract_dir)

    elif choice == "0":
        break
    else:
        print("Lá»±a chá»n khÃ´ng há»£p lá»‡!")
```
